```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
url="https://coinmarketcap.com/"
r= requests.get(url)
htmlcontent=r.content
htmlcontent
```

```python
soup = BeautifulSoup(htmlcontent, 'html.parser')
print(soup.prettify)
```

```python
cnames = soup.find_all(class_="sc-71024e3e-0 ehyBa-d")
cnames
```

```python
name=[]
for i in cnames:
    name.append(i.get_text())

print(name)

len(name)
```

```python
symb=soup.find_all(class_="sc-71024e3e-0 OqPKt coin-item-symbol")
symb
```

```python
symbol=[]
for i in symb:
    symbol.append(i.get_text())

print(symbol)

len(symbol)
```

```python
rs=soup.find_all(class_="sc-a093f09c-0 gPTgRa")
rs
```

```python
b=[]
for i in rs:
    b.append(i.get_text())

print(b)

len(b)
```

```python
import re
price = [re.sub(r'[\$,]', '', value) for value in b]
print(price)
```

```python
hour=soup.find_all(class_="sc-a59753b0-0 ivvJzO")
hour
```

```python
one_h=[]
for i in hour:
    one_h.append(i.get_text())

print(one_h)

len(one_h)
```

```python
one_h=one_h[0:10]
one_h

len(one_h)
```

```python
one_day=soup.find_all(class_="sc-a59753b0-0 ivvJzO")
one_day
```

```python
day1=[]
for i in one_day:
    day1.append(i.get_text())

print(day1)

len(day1)
```

```python
day1=day1[0:10]
day1
```

```python
a=[]
for i in cap:
    a.append(i.get_text())

print(a)

len(a)
```

```python
import re
market_cap = [re.sub(r'[\$,]', '', value) for value in a]
print(market_cap)
```

```python
vol=soup.find_all(class_="sc-71024e3e-0 bbHOdE font_weight_500")
vol
```

```python
import pandas as pd
Cryptocurrency = pd.DataFrame({
    "Name": name,
    "Symbol": symbol,
    "Price": price,
    "One_hour":one_h,
    "One_Day":day1,
    "Market_Cap": market_cap,
    "Volume":volume,
    "Volume2":volume1,
    "Circulating_Supply":supply
})
Cryptocurrency
```

```python
Cryptocurrency.to_csv("Cryptocurrency_Data_Extraction.csv",index=False)
```
